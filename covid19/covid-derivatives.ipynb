```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import csv

dates = np.array([])
cases = np.array([])

with open('usa-covid19.csv') as csvDataFile:  #data current through 3/26/2020
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        dates = np.append(dates,row[0])
        cases = np.append(cases,float(row[1]))

plt.plot(cases,'bo-');
```

```python
#1st derivatives many ways

n = len(cases)
h=1 #step is 1 day
fdcd = np.zeros(n)
fdos3 = np.zeros(n)

for j in range(1,n-1):
    fdcd[j] = (cases[j+1]-cases[j-1])/(2.*h)

for j in range(2,n):
    fdos3[j] = (3.*cases[j] - 4.*cases[j-1] + cases[j-2])/(2.*h)

plt.plot(fdcd[-21:-1],'r.',fdos3[-21:n],'bo');
plt.legend(['center differencing','one-sided differencing']);
```

```python
#2nd derivatives many ways

n = len(cases)
h=1 #step is 1 day
sdcd = np.zeros(n)
sdos4 = np.zeros(n)

for j in range(1,n-1):
    sdcd[j] = (cases[j-1]-2*cases[j]+cases[j+1])/(h*h)

for j in range(3,n):
    sdos4[j] = (2.*cases[j] -5.*cases[j-1] + 4.*cases[j-2] - cases[j-3])/(h*h)

plt.plot(sdcd[1:-1],'r.',sdos4[3:n],'bo');#,fdos3[-21:-2],'b:');
plt.legend(['center differencing','one-sided differencing']);
```
